# 2.2.1 Probability Axioms



The probability axioms are fundamental rules for probability. They state that for any event its probability must be greater than or equal to 0, the sum of all mututal exclusive events in a sample space is 1, and that the union of two independent events is the sum of their probabilities.

```python
import numpy as np

probabilities = np.array([0.2, 0.5, 0.3])

# Check non-negativity
non_negative = np.all(probabilities >= 0)
print("Non-negativity check:", non_negative)  # this should print true since they are all above 0

# Sum of probabilities
total_probability = np.sum(probabilities)
print("Total probability (should be 1):", total_probability)  # Should print 1.0

# We will create the independent events A and B
P_A = 0.5
P_B = 0.4

# for independent evens, P(A or B) = P(A) + P(B)
# this is the probability that either event A or B happens
P_A_or_B = P_A + P_B
print("P(A) + P(B) =", P_A_or_B)  # Should be 0.7

# For independent events, P(A and B) = P(A) * P(B)
#This is the probability that both event A and B will happen at the same time
P_A_and_B = P_A * P_B
print("P(A and B) =", P_A_and_B)  # Should print 0.2

```

# 2.2.2 Conditional Probability

This is the probability of an event A happening given that event B already happened. This is represented by the folowing notation $P(A | B)$.

This can be calculated with the following equation:

$P(A | B) = \frac{A \cap B}{P(B)}$




```python
#We will do an example with a deck of cards

total_cards = 52

face_cards = 12

queens = 4

#Event A is drawing a queen
#event B is drawing a face card

#Our conditional probability will be that given you drew a face card what is the
#what is the probability it is  aqueen

P_A_and_B = queens / total_cards #the intersection of A and B is that a queen is drawn
P_B = face_cards / total_cards
P_A_given_B = P_A_and_B / P_B # this is like the formula shown above


print("P(A and B) (drawing a queen):", P_A_and_B)
print("P(B) (drawing a face card):", P_B)
print("P(A | B) (drawing a queen given it's a face card):", P_A_given_B)
```

# 2.2.3 Discrete Random Variables

Discrete means countable so they cannot take on values in between.

Discrete random variables have finite values. They also can be described by a probability mass function or PMF which assigns a probability to each possible discrete value.

An example of something with a discrete random variable is a dice roll. The PMF of a dice roll is

```python
import numpy as np
import matplotlib.pyplot as plt

# The outcomes for a 6 sided die are 1 to 6
outputs = np.array([1, 2, 3, 4, 5, 6])

pmf = np.array([1/6] * 6) # the probability mass function is 1/6 for each because
# each side has an equal probability

# This will show the probability of each outcome
for outcome, probability in zip(outputs, pmf):
    print(f"P(X = {outcome}) = {probability:.4f}")

# We will plot the outcomes and probabilites found above
plt.bar(outputs, pmf)
plt.xticks(outputs)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function of a Six-Sided Die')
plt.ylim(0, 1)
plt.show()
#As you can see they are all very similar which shows how its also uniform
```

# 2.2.4 Continuous Random Variables

A continuous random variable can take on an **infinite** number of possibe values within a given range. Its often used for measurments or quantities that can be divided into smaller increments like a measurment.

A continuous random variable uses a probability density function instead of a probability mass function. The probability density function gives the probability of a variability falling within a range. The cumulative densitg fuction gives an accumulated sum of all of the values up to a certain value.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
#The following example will be for heights
# Parameters for the normal distribution
mean = 175  # average height in cm
std_dev = 7  # standard deviation in cm

# Generate synthetic data: heights of 1000 people
num_samples = 1000
heights = np.random.normal(mean, std_dev, num_samples)

# Make a range of values using the mean and stdev
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# calculte the probability density function
pdf = stats.norm.pdf(x, mean, std_dev)

# Calculate the cumulative density function
cdf = stats.norm.cdf(x, mean, std_dev)

# Plotting uthe PDF
plt.figure(figsize=(12, 5))

# PDF plot
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.hist(heights, bins=30, density=True, alpha=0.5, color='gray')
plt.title('Probability Density Function (PDF) of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.legend()

# CDF plot
# Notice how it goes up to 1
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='green')
plt.title('Cumulative Distribution Function (CDF) of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.show()
```
