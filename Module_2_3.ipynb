# 2.3.1 Joint probability distributions

This gives the probabilty that two or more random variables take on specific values simultaneously. It describes the likelihood of different combinations of outcomes for those variables.

If $X$ and $Y$ are discrete random variables then the joint probability mass function is given by $P(X = x, Y = y) = p(x,y)$ This function takes the probability that $X$ has the value $x$ and Y has the value $y$ simultaneously.


```python
import numpy as np

#We have two discrete random variables, X and Y
#In this situation they are both the rolling of a 6 sided dice.
X_3 =  1/6 #This is the probability of X = 3
Y_4 = 1/6 #This is the probability of X = 4

jointProbability = X_3 * Y_4
print(f"Joint probability P(X=3, Y=4): {jointProbability}") # or 1/36
```

# 2.3.2. Correlation and dependence

Both of these terms are used to describe the relationship between variables.

**Dependence** is a concept that describes the statistical relationship between two or more variables where knowing the value of one variable provides insights into the value of another variable. It can be linear, non linear, or stochastic.

Two variables are independent if they do not affect each other. Below is an example of independence:

$P(x = x, Y = y) - P(X = x)P(Y=y) $

For example, flipping two coins is independent because the first coin flip does not affect the second one.

Correlation a type of dependence that is linear. Therefore, any two variables that are correlated are dependent. An example of correlation could be the height and weight of people, where as the height increases, the weight does as well.


```python
def plotPoints(plot,X,Y,color,title,x_label,y_label):
  plot.scatter(X, Y, color=color, label=title)
  plot.set_title(title)
  plot.set_xlabel(x_label)
  plot.set_ylabel(y_label)

```

```python
import matplotlib.pyplot as plt
n_sample_size = 100

# First we will show linear correlation

x1 = np.random.randn(n_sample_size);
y1 = 2 * x1 + np.random.randn(n_sample_size) * 0.2 #we will use this to add noise
# Next we will show Non-Linear Dependence
x2 = np.random.uniform(-2, 2, n_sample_size)
y2 = x2 ** 2 + np.random.randn(n_sample_size) * 0.2

# Finally we will show no dependence
x3 = np.random.randn(n_sample_size)
y3 = np.random.randn(n_sample_size)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

plotPoints(ax[0],x1,y1,'b',"Linear Dependence","X1","Y1")
plotPoints(ax[1],x2,y2,'g',"Non-Linear Dependence","X2","Y2")
plotPoints(ax[2],x3,y3,'r',"No Dependence","X3","Y3")
plt.tight_layout()

#Look at the plots below, notice the correlation for the linear one
#Then look at the non linear depdence plot, it has a quadratic structure
#The non dependnt plot has no pattern

```

# 2.3.3. Random samples

A random sample is a subset of data points taken from a larger population. Typically, each member of the population has a random chance of being selected. There also exists different types of random sampling:

1. Simple Random Sampling
    * Every member has an equal chance of being included
2. Stratified Sampling
  *  The population is divided into subgroups or strata and then a random sample is taken from each group
3. Cluster Sampling
  * The population is divided into clusters which are groups and then groups are randomly selected
4. Systematic Sampling
  * Every nth member of the population is chosen after a random starting point.

```python
import numpy as np

# Our population is the number 1 -100
population = np.arange(1, 101)

# We want to have a sample of 10 numbers

#Simple Random Sampling
sample = np.random.choice(population, size=10, replace=False)
print("Random sample:", sample)
#Stratified Sampling
# We will split the sample into two stratas (even & odd numbers)
even_numbers = population[population % 2 == 0]
odd_numbers = population[population % 2 != 0]
#We will pick 5 from each strata
sample_even = np.random.choice(even_numbers, size=5, replace=False)
sample_odd = np.random.choice(odd_numbers, size=5, replace=False)

stratified_sample = np.concatenate([sample_even, sample_odd])
print("Stratified Sample:", stratified_sample)

#Cluster Sampling
clusters = np.split(population, 10)
chosen_index = np.random.choice(len(clusters), size=1, replace=False)
cluster_sample = clusters[chosen_index[0]]
print("Cluster Sample:", cluster_sample)
#Systematic Sampling
#We will select every 10th element
k = 10
systematic_sample = population[::k]
print("Systematic Sample:", systematic_sample)
```
