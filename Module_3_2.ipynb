# 3.2.1. Limits and continuity
## Limits
These are both concepts used to describe the behavior of a function as they behave near a point or within an interval.

A limit describes the value of a function as it approaches a certain value.

The limit of a function f(x) as x approaches c is denoted as:
$ \lim_{x \to c}f(x) = L$

This means that as $x$ gets closer and closer to $c$ the function $f(x)$ approaches the value $L$.

### Example 1
If $f(x) = 2x$, then the limit of $f(x)$ as $x$ approaches 3 is:
$\lim_{x\ \to 3}2x = 6$

Essentially, as $x$ gets closer to 3, $f(x)$ approaches 6.
## Continuity

A function is said to be continuous at a point if there are no breaks, jumps, or holes in its graph at that point. A function $f(x)$ is continuous at a point $c$ is the three following conditions are met:


1.   The function is well-defined at $c$
2.   The limit of the function as $x$ approaches $c$ exists
3.   The limit of the function as $x$ approaches $c$ equals the value of that function at that point: $\lim_{x \to c} f(x) = f(c)$

### Example 2
Consider the function $f(x) = x^2$. We can show that this function is continuous at $x = 2$ becasuse


*   The function is defined at $x = 2$ as $f(2) = 4$
*   The $\lim_{x \to 2} x^2 = 4$, the limit exists
*   $\lim_{x \to 2} x^2 = 4 = f(2)$, the limit equals the value of the function

If any of these conditions are not met by the function then the function is said to be discontinuous at that point.

A function is said to be continuous at all points when
the function is continuous at every point.

# 3.2.2. Derivatives
The derivative of a function represents the rate at which the function's value changes as its input changes. Essentially it measures the rate of change of the output of a function in relation to a unit change in the input.

## Basic Definition of a Derivaitve:
The derivative of a function $f(x)$ at a point $x = a$ is defined as the limit of the average rate of change:

$f'(a) = \lim_{h \to 0}\frac{f(a + h) - f(a)}{h}$
In the expression above, $f'(a)$ is the derivative of $f(x)$ at $x= a$
## Interpretation of a Derivative
Within physics, the derivative of a position is velocity, and the derivative of speed is acceleration.

Geometrically or on a plot its the slope of the tangnet line of a curve at a point.

If a derivative is above 0, then the function is increasing.

If a derivative is below 0, then the function is decreasing.

## Applications of a Derivative
Within physics derivatives can be used to determine a rate of change.

Within economics, derivatices can represent marginal cost or marginal revenue.

With biology, derivatives can model rates of growth or decay.

There are other applications as well.

## Higher Order Derivatives
A higher order derivative is essentially a derivative of a derivative.

A second order derivative, denoted $f"(x)$ is a derivative of the first derivative.


```python
# Let's manually compute the derivatives using numpy and plot them

#We can compute the derivatives using the power rule
# f(x) = x^3 + 2x^2 + x + 7
def f(x):
    return x**3 + 2*x**2 + x + 7

# f'(x) = 3x^2 + 4x + 5
def f_prime(x):
    return 3*x**2 + 4*x + 1

# f''(x) = 6x + 4
def f_double_prime(x):
    return 6*x + 4

# Generate x values for plotting
x_vals = np.linspace(-10, 10, 400)

f_vals = f(x_vals)
f_prime_vals = f_prime(x_vals)
f_double_prime_vals = f_double_prime(x_vals)

plt.figure(figsize=(10, 6))

plt.plot(x_vals, f_vals, label=r'$f(x) = x^3 + 2x^2 + 1x + 7$', color='blue')
plt.plot(x_vals, f_prime_vals, label=r"$f'(x) = 3x^2 + 4x + 5$", linestyle='--', color='green')
plt.plot(x_vals, f_double_prime_vals, label=r"$f''(x) = 6x + 4$", linestyle=':', color='red')

# Adding titles and labels
plt.title('Plot of a Function and its Derivatives')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.legend()

plt.show()
```

# 3.2.3. Taylor's theorem

The taylors theorem is a way to approximate a fucntion near a given point using a polynomial. The function uses the function's derivatives to form the polynomial. The more terms in the polynomial then the more accurate the approximation is.

Given a function $f(x)$ that is infinitely differentiable, we can approximiate it with a Taylor polynomial.

The Taylor Series is how you can create this polynomial:
$f(x) = f(a) = f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \frac{f^3(a)}{3!}(x-a)^3 +...$

This can be represented by the following series
$f(x) = ∑_{n = 0}^∞\frac{f^n(a)}{n!}(x-a)^n$

$f^n(a)$ is the nth derivative of $f(x)$ at $x= a$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
x0 = 2  # center of expansion
x = np.linspace(-5, 10, 1000)  # Range of x values
y_actual = x**2  # the Actual function values

# Zero order Taylor approximation
T0 = x0**2 * np.ones_like(x)

# First order Taylor approximation
T1 = x0**2 + 2 * x0 * (x - x0)

# Second order Taylor approximation
T2 = x0**2 + 2 * x0 * (x - x0) + 1 * (x - x0)**2

plt.figure(figsize=(12, 6))
plt.plot(x, y_actual, label='$x^2$', color='black', linewidth=2)

plt.plot(x, T0, label='Zero order Taylor Approximation', linestyle='--', color='red')
plt.plot(x, T1, label='First order Taylor Approximation', linestyle='--', color='green')
plt.plot(x, T2, label='Second order Taylor Approximation', linestyle='--', color='blue')

plt.title('Taylor Series Approximation of $x^2$ around $x_0 = 2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
