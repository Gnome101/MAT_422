# 3.4. Logistic regression

Logistic regression is a statistical method for predicting finite outcomes  based on one or more predictor variables.

A predictor variable also known as an independent variable is used to predict or explain the outcome of another variable.

The function for logistic function is also called the sigmoid function looks like this:

$\sigma(z) = \frac{1}{1+e^{-z}}$

$z = β_0 + β_1x +β_2x_2+ … +β_nx_n​$

$x_1, x_2, …, x_n $ are the predictor variables

$\beta$ is represents the model coefficients or parameters it essentially describes the strength and direction of the relationship between the predictor variable and the response variable.

$β_0$ is the intercept term which is the output when all else is 0.

The logistic function essentially squashes the output of $z$ into the range $[0,1]$.

# Applications

Logistic regression is used in many fields because it provides a simple way to model binary outcomes. In healthcare it is used to determine if a patient has a disease, in finance it can be used to determine fraudulent transactions. It can also be used for spam detection to determine if an email is spam or not.

```python
import numpy as np
import matplotlib.pyplot as plt



# Coefficients
beta_0 = -3  # intercept or the baseline value
beta_1 = 0.8  # coefficient for predictor X_1

X = np.linspace(0, 50, 1000)

z = beta_0 + beta_1 * X

probabilities = 1 / (1 + np.exp(-z)) #this is the sigmoid function

plt.figure(figsize=(10, 6))
plt.plot(X, probabilities, label=r'$P(Y=1 | X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}$')
plt.title('Logistic Regression: Probability vs. Predictor Variable')
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Probability of Outcome (P(Y=1 | X))') # The equation is a conditional probability
plt.axhline(0.5, color='grey', linestyle='--', label='Threshold (P=0.5)')
plt.legend()
plt.grid(True)
plt.show()
# From this we can see that a higher X value makes the outcome more likely.
```
