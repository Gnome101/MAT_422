# 4.2. Graph and graph modeling
A graph is a mathematical structure used to model relationships between objects, its widely used in computer science, discrete mathematics, network analysis, and other fields.

A graph is made of two components:
A vertice and edge. A vertice is also an object, and they are connected to each other edge. A graph can be represented formally as $G = (V,E)$ where $G$ is the graph, $V$ is a set of vertices, and $E$ is a set of edges where each edge is a pair of vertices represented like $(u,v)$.

## Types of Graphs
There exist directed graphs which have a direction and exist as ordered pairs. There are also undirected graphs which have unordered pairs with bidirectional connections.

Graphs can also have edges with weights or costs associated with them.

## Graph Representation
A graph can be represted using an adjacency list
This is a list of lists where each vertex has a list of adjacent vertices.
An adjacency matrix is a 2D array where A[i][j] indicates the presence or weights of an edge
An edge list is a list of all edges.

## Graph Modeling
This is creating a graph represntation of a problem to try and study or solve it. First you need to identify the entities (vertices) and their relationships(edges) in the system. Then to analyze the graph you can look at the shortest path, connectivity, cycle detection, etc.

```python
#We can install networkx to create and work with our graph easily
pip install networkx
```

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.DiGraph()

# We can add the following edges and weights
edges = [
    ('A', 'B', 2),
    ('A', 'C', 4),
    ('B', 'D', 1),
    ('C', 'D', 3),
    ('D', 'E', 5)
]
G.add_weighted_edges_from(edges)

print("Adjacency List:")
for line in nx.generate_adjlist(G):
    print(line)

print("\nEdge List:")
for u, v, weight in G.edges(data='weight'):
    print(f"({u}, {v}, weight={weight})")

print("\nBFS from 'A':", list(nx.bfs_edges(G, source='A')))
print("DFS from 'A':", list(nx.dfs_edges(G, source='A')))

print("\nShortest Path 'A' to 'E':", nx.shortest_path(G, source='A', target='E', weight='weight'))
print("Shortest Path Length 'A' to 'E':", nx.shortest_path_length(G, source='A', target='E', weight='weight'))

# Visaulze the graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # spring_layout returns the position for all of the nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): w for u, v, w in edges})
plt.title("Graph Visualization")
plt.show()
```
